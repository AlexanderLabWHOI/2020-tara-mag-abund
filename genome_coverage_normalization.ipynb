```python
import pandas as pd
import os
import glob
```

```python
initial_df = pd.read_csv('output/ERR1726699.coverm.abundance.tab', sep='\t', index_col=0)
initial_df.columns = initial_df.columns.str.split(' ').str[1:].str.join('_')
```

```python
lengths = initial_df['Length'].drop('unmapped')
```

```python
all_reads = pd.DataFrame(index = initial_df.index)
for f in glob.glob('output/*tab'):
    name = os.path.basename(f).split('.')[0]
    if name == 'ERR1726706':
        pass
    else:
        tt = pd.read_csv(f, sep='\t', index_col=0)
        tt.columns = tt.columns.str.split(' ').str[1:].str.join('_')
        all_reads[name] = tt['Read_Count']
all_reads= all_reads.drop('unmapped')
```

```python
read_counts = pd.DataFrame(columns = ['all_reads', 'trimmed_surviving'])
with open('trimmed-read-stats', 'r') as f:
    for line in f:
        l=line.split(':')
        name = l[0].split('.')[0]
        all_reads = int(l[2].strip().split(' ')[0])
        trimmed_reads_surving_pairs = int(l[3].split(' ')[1])
        read_counts.loc[name] = [all_reads, trimmed_reads_surving_pairs]
```

```python
library_size = read_counts['trimmed_surviving']
```

```python
summed_reads = all_reads.sum()
```

Equations from https://haroldpimentel.wordpress.com/2014/05/08/what-the-fpkm-a-review-rna-seq-expression-units/

X = total reads recruiting to a genome

l = length of genome in Kb

N = total number of trimmed reads in millions

\begin{align}
RPKM_{i} & = \frac{X_i}{l_i  N } * 10^9
\end{align}

```python
rpkm = (all_reads).div(library_size).div(lengths, axis=0)*1e9
```

TPM is normalized to all total RPKM
\begin{align}
TPM_{i} & = \frac{RPKM_i}{\sum RPKM} * 10^6
\end{align}

```python
tpm = rpkm/rpkm.sum()*1e6
```

```python
tpm.to_csv('MAG_tpm.csv')
```
